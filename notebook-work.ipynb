```python
train_new = train.fillna(-999)
test_new = test.fillna(-999)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
le=LabelEncoder()
train_new['hospital_admit_source'] = le.fit_transform(train_new['hospital_admit_source'])
train_new['icu_admit_source'] = le.fit_transform(train_new['icu_admit_source'])
train_new['ethnicity'] = le.fit_transform(train_new['ethnicity'])

train_new['gender'] = le.fit_transform(train_new['gender'])
train_new['icu_type'] = le.fit_transform(train_new['icu_type'])
train_new['icu_stay_type'] = le.fit_transform(train_new['icu_stay_type'])
```

```python
test_new['hospital_admit_source'] = le.fit_transform(test_new['hospital_admit_source'])
test_new['icu_admit_source'] = le.fit_transform(test_new['icu_admit_source'])
test_new['ethnicity'] = le.fit_transform(test_new['ethnicity'])

test_new['gender'] = le.fit_transform(test_new['gender'])
test_new['icu_type'] = le.fit_transform(test_new['icu_type'])
test_new['icu_stay_type'] = le.fit_transform(test_new['icu_stay_type'])
```

```python
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier
import catboost
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
seed = 1080
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
scores = []
preds = []
model=CatBoostClassifier(verbose=False)

#creating a for loop for the stratified k fold
i = 1
for train, test in skf.split(X, y):
    x_train, x_test, y_train, y_test = X.iloc[train], X.iloc[test], y.iloc[train], y.iloc[test]
    model.fit(x_train, y_train)
    print('Number of splits trained {} '.format(i))
    score = roc_auc_score(y_test,model.predict(x_test))
    pred = model.predict_proba(test_new)[:, 1]
    scores.append(score)
    preds.append(pred)
    i += 1

np.mean(scores)
```
